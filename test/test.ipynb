#### test rating regex

```python
import re

# List of levels
levels = ['A+', 'A', 'B', 'S']

# Input string to search
input_string = "The grade for this assignment is A+."

# Create a regular expression pattern using the elements in the list
pattern = '|'.join(re.escape(level) for level in sorted(levels, key=len, reverse=True))

# Use re.search to find a match in the input string
match = re.search(pattern, input_string)

# Check if there is a match
if match:
    matched_level = match.group()
    print(f"Matched level: {matched_level}")
else:
    print("No match found")
```

#### test different doc loaders

```python
from langchain.document_loaders.word_document import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
doc_file_path = "../data/raw/cnki_papers_zh_mini/人工智能时代学生数据隐私保护的动因与策略_侯浩翔.docx"
```

```python
docx2txt_loader = Docx2txtLoader(doc_file_path)
uns_doc_loader = UnstructuredWordDocumentLoader(doc_file_path)

# if in mode 'elements', the unstructured library will split the document into elements such as Title and NarrativeText.
uns_ele_doc_loader = UnstructuredWordDocumentLoader(doc_file_path, mode="elements", strategy='fast')
```

```python
docx2txt = docx2txt_loader.load()
uns_doc = uns_doc_loader.load()
uns_ele_doc = uns_ele_doc_loader.load()
```
